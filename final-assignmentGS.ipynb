```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
defaulters = df[df['default'] == 1]

X_defaulters = defaulters.drop(columns=['ID', 'default'])
X_defaulters['SEX'] = X_defaulters['SEX'].map({'M': 0, 'F': 1})
X_defaulters = pd.get_dummies(X_defaulters, columns=['EDUCATION', 'MARRIAGE'], dtype=int)
```

```python
scaler = StandardScaler()
X_defaulters_scaled = scaler.fit_transform(X_defaulters)
```

```python
inertia = []
k_range = range(2, 10)
for k in k_range:
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(X_defaulters_scaled)
    inertia.append(km.inertia_)

plt.figure(figsize=(8, 4))
plt.plot(k_range, inertia, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

```python
km = KMeans(n_clusters=4, random_state=42)
clusters = km.fit_predict(X_defaulters_scaled)

defaulters = defaulters.copy()
defaulters['cluster'] = clusters
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_defaulters_scaled)
```

```python
plt.figure(figsize=(8, 5))
for i in range(4):
    mask = clusters == i
    plt.scatter(X_pca[mask, 0], X_pca[mask, 1], label=f'Cluster {i}', alpha=0.5)
plt.title('Defaulter Clusters')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend()
plt.show()
```

```python
profile = defaulters.groupby('cluster')[['LIMIT_BAL', 'PAY_0', 'BILL_AMT1', 'PAY_AMT1', 'AGE']].mean().round(2)
print(profile)
print(defaulters['cluster'].value_counts())
```

```python
profile.drop(index=3).plot(kind='bar', figsize=(10, 6))
plt.title('Average Feature Values by Defaulter Cluster')
plt.xticks(rotation=0)
plt.show()
```

```python
features_to_plot = ['LIMIT_BAL', 'PAY_0', 'BILL_AMT1', 'PAY_AMT1', 'AGE']
profile_clean = profile.drop(index=3)

fig, axes = plt.subplots(1, 5, figsize=(18, 5))
for i, feature in enumerate(features_to_plot):
    profile_clean[feature].plot(kind='bar', ax=axes[i], color='steelblue')
    axes[i].set_title(feature)
    axes[i].set_xlabel('Cluster')
    axes[i].tick_params(axis='x', rotation=0)

plt.suptitle('Defaulter Cluster Profiles', fontsize=14)
plt.tight_layout()
plt.show()
```
